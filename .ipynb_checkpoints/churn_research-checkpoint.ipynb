# II. LSTM with multiple multivariate time series

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense

# B : false churn
# C : false retain
# D : true churn

def recall(true, predicted):
    mult = true*predicted
    sub = true-predicted
    D = sum(mult)
    C = sum(sub == 1)
    return round(D/(C+D),2)

def precision(true, predicted):
    mult = true*predicted
    sub = true-predicted
    D = sum(mult)
    B = sum(sub == -1)
    return round(D/(B+D),2)

def f1(true, predicted):
    r = recall(true,predicted)
    p = precision(true,predicted)
    return round(2*r*p/(r+p),2)
```
